```python
## librerias necesarias para el webscrapping ##
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium_stealth import stealth
from selenium.common.exceptions import TimeoutException, StaleElementReferenceException
from datetime import datetime
import time, os 
from selenium.webdriver.common.alert import Alert
#############################################

```

## precargar proceso ##

```python
#############################################

# Creacion de cliente de automatizador #

def registro_masivo(usuario,clave,name_list,person, ruta):

    #inicializar el servicio chrome automatizacion 

    service = Service(ChromeDriverManager().install())
    option = webdriver.ChromeOptions()

    ### mostrando pantalla del funcionamiento ###

    option.add_argument("--window-size=1920,1080")

    ###sin mostrar pantalla ###

    #option.add_argument("--headless") 

    ### desabilita las extenciones (agiliza el funcionamiento) ###

    option.add_argument("--disable-extensions")

    ### desabilita el reconocimento de webdriver ###

    option.add_argument("--disable-blink-features-AutomationControlled")

    ### desactiva los parametros de automatizacion para hacer pasar selenium como humano ###

    option.add_experimental_option("excludeSwitches", ["enable-automation"])
    option.add_experimental_option('useAutomationExtension', False)
    driver= Chrome(service=service, options = option)

    ## funcion que simula interfaz humana ##
    stealth(driver,
        languages=["en-US", "en"],
        vendor="Google Inc.",
        platform="Win32",
        webgl_vendor="Intel Inc.",
        renderer="Intel Iris OpenGL Engine",
        fix_hairline=True,
        )

################################################################

    # ejecucion del servicio #
   
    ### inicializar navegador ###
    website = "https://aulamatriz.edu.co/correo_masivo/" # url
    driver.get(website)

    ### proceso de creacion de lista ###

    #   iniciar sesion en la pagina de consulta #
    user = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/div/div/div/div/form/div[1]/input"))) 
    user.send_keys(usuario)

    ## ingresar clave ###
    password = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/div/div/div/div/form/div[2]/input"))) 
    password.send_keys(clave)

    ## click en iniciar sesion ###
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/div/div/div/div/form/div[3]/input"))).click() 

    ## click en crear lista ##
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"    /html/body/div/div/nav/ul/li[8]/div/ul/li[1]/a"))).click() 

    ## llenar formulario lista ##
    nombre_lista = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/div/div/div/div/div/div/div/div[1]/div/div/form/div[1]/input"))) 
    nombre_lista.send_keys(name_list)

    ## seleccionar pesona encargada ##
    DropdewonDocumento = Select(WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/div/div/div/div/div/div/div/div[1]/div/div/form/div[2]/select"))))
    DropdewonDocumento.select_by_value(person)

    ## crear la descripcion de la lista ##
    descripcion = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/div/div/div/div/div/div/div/div[1]/div/div/form/div[3]/textarea"))) 
    descripcion.send_keys("Divulgacion")

    ## click en crear lista ##
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/div/div/div/div/div/div/div/div[1]/div/div/form/input"))).click() 
    

    WebDriverWait(driver, 10).until(EC.alert_is_present())
    driver.switch_to.alert.accept()
    ### proceso de creacion de lista terminado ###
    ## guardar id de la lista  ##
    id = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/div/div/div/div/div/div/div/div[1]/div/div/div/table/tbody/tr[1]/td[1]")))
    id = id.text

    ### proceso de administracion de lista ###
    ## cargue de la BD de lista ##
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/div/div/div/div/div/div/div/div[1]/div/div/div/table/tbody/tr[1]/td[6]/a[1]"))).click() 
    
    ## seleccion de bd ##
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/div/div/div/div/div/div/div/div[1]/div/div/div/ul/li[2]/a"))).click() 
    ## cargar elarchivo de la BD ##
    # Localiza el elemento de tipo "file" y env√≠a el archivo
    archivo_input = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "/html/body/div/div/div/div/div/div/div/div/div/div[1]/div/div/div/div[2]/form/div/div[1]/div/input")))
    archivo_input.send_keys(ruta)  # Ruta del archivo a cargar
    ## click en cargar archivo ##
    
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/div/div/div/div/div/div/div/div[1]/div/div/div/div[2]/form/div/div[2]/input"))).click()
    WebDriverWait(driver, 10).until(EC.alert_is_present())
    driver.switch_to.alert.accept()
   
    time.sleep(10)
    
    ## crear nuevo envvio ##
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/nav/ul/li[10]/a/span"))).click() 
    time.sleep(5)
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/nav/ul/li[10]/div/ul/li[1]"))).click() 
    
    ## seleccionar mensaje ista ##
    nombre_lista = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/div/div/div/div/div/div/div/div[1]/div/div/form/div[1]/div/div/div/div[1]/input"))) 
    nombre_lista.send_keys(name_list)
        
    ## seleccionar lista ##
    DropdewonLista = Select(WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/div/div/div/div/div/div/div/div[1]/div/div/form/div[1]/div/div/div/div[1]/select"))))
    DropdewonLista.select_by_value(id)

    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/div/div/div/div/div/div/div/div[1]/div/div/form/div[1]/div/div/button"))).click() 
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH,"/html/body/div/div/div/div/div/div/div/div/div/div[1]/div/div/form/div[2]/div[1]/div[1]/input"))).click() 
    
    time.sleep(120)
    try :
        driver.close()
        driver.quit()
    except:
        return "realizada la operacion, pero usted cerro la ventana"

```
